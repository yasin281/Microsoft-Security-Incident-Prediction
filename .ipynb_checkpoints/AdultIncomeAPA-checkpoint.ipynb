# Evaluación Preliminar del Conjunto de Datos: Adult Dataset

Este notebook presenta un análisis básico del conjunto de datos `adult.data`. En esta evaluación preliminar, se describe el conjunto de datos, se identifica la variable objetivo y el tipo de problema, y se visualizan las relaciones entre las variables. También se incluyen la detección de posibles problemas con los datos y la matriz de correlación para evaluar interdependencias.


```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

from time import time
from datetime import timedelta
init_time = time()

import pandas as pd
import numpy as np
import seaborn as sns
```

```python


column_names = ["age", "workclass", "fnlwgt", "education", "education-num", 
                "marital-status", "occupation", "relationship", "race", "sex", 
                "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]

data = pd.read_csv("/home/momin/Downloads/census+income/adult.data", names=column_names, na_values=" ?", skipinitialspace=True)

#los missing values estan representados por '?'
data = data.replace('?', np.nan)

data.head()

```

## Descripción del Conjunto de Datos

Este conjunto de datos contiene información demográfica y laboral de individuos, con el objetivo de predecir si el ingreso anual es superior o inferior a 50K. La fuente de los datos es el repositorio de UCI Machine Learning. La variable objetivo es **income**, y este es un problema de **clasificación**.

A continuación, se presenta una breve descripción de las características principales y algunos posibles problemas a considerar.


```python
data.info()
```

## Posibles Problemas en el Conjunto de Datos

1. **Valores Faltantes**: Algunos atributos tienen valores faltantes representados por "?" que hemos reemplazado por NaN.
2. **Desbalance entre Clases**: Verificaremos la distribución de la variable objetivo `income` para analizar si las clases están equilibradas.


```python
import matplotlib.pyplot as plt 

missing_values = data.isnull().sum()
print("Valores nulos en el dataset:\n", missing_values)

plt.figure(figsize=(8, 6))
sns.countplot(data=data, x="income")
plt.title("Distribución de la Variable Objetivo ('income')")
plt.xlabel("Income")
plt.show()
```

## Visualización de Variables en Relación con el Objetivo

A continuación, se muestran visualizaciones simples de cómo las variables se relacionan con el objetivo para identificar posibles patrones y relaciones.


```python
numeric_features = data.select_dtypes(include=['float64', 'int64']).columns

for feature in numeric_features:
    if feature != "income":
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=data, x="income", y=feature)
        plt.title(f"{feature} vs Income")
        plt.xlabel("Income")
        plt.ylabel(feature)
        plt.show()
```

## Matriz de Correlación

La matriz de correlación muestra la relación entre las variables numéricas. Esto nos ayudará a identificar características altamente correlacionadas que pueden influir en el análisis o el modelado.


```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_data.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

# Evaluación Preliminar con Modelo de Referencia: Regresión Lineal

Este notebook implementa un modelo simple de **Regresión Lineal** para predecir la variable `income` en el conjunto de datos `adult.data`. Este modelo se usa como referencia para evaluar la complejidad del conjunto de datos y decidir si es adecuado para nuestro proyecto.


```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt

column_names = ["age", "workclass", "fnlwgt", "education", "education-num", 
                "marital-status", "occupation", "relationship", "race", "sex", 
                "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]

data = pd.read_csv("/home/momin/Downloads/census+income/adult.data", names=column_names, skipinitialspace=True)

data = data.replace('?', np.nan).dropna()

data['income'] = data['income'].apply(lambda x: 1 if x == '>50K' else 0)

data = pd.get_dummies(data, drop_first=True)

X = data.drop('income', axis=1)
y = data['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R2 del modelo de Regresión Lineal: {r2:.2f}")
```
